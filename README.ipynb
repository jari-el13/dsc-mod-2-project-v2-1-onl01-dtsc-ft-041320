# Motivation
- For this project we were trying to put a model together to predict housing prices in King County.  King County is located in Seattle, Washington.  The model that was used is Multiple Linear Regression.  We used this in order to see what features would have the strongest sway on the prices in order to get an acurate prediction of hosuing prices.

# Objective
- We want to find the features that would help maximize profit if one is looking to sale and the best deal for one looking to buy.

# Recommendations
- Buyers
    - The time period the house was built in should not have an impact on your decison as far as price.  There are lots of old houses that are just as expensive as new houses.  This indicates there have been referbs.
    - Also if you are thinking of living near the beach of Lake Washington in the Bellevue area or Downtown Seattle then you should expect to pay more as these are the most expensive places to buy a house.


- Seller
    - There is strong correlation with how large the house is and how much it cost in King's County.  So try and buy large houses cheap and sell them for high.
    - Also don't let the year the house was built deter you from keeping all options open.  There are houses built in the 1900's that are worth the same as houses built as new as the 2000's if not more!
